# **1. Memasukan Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset/avocado.csv')

# Menampilkan lima baris pertama dari dataframe
print(df.head())
```

# **3. Memisualisasikan Data**

### **Box Plot**
ChatGPT Prompt: Beri kode untuk menampilkan boxplot dari fitur - fitur numerik

```python
# Memilih kolom-kolom numerik
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Menggambar boxplot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
sns.boxplot(data=numeric_columns)
plt.xticks(rotation=45)  # Mengatur rotasi label sumbu x agar lebih mudah dibaca
plt.title('Boxplot untuk Fitur Numerik')
plt.xlabel('Fitur')
plt.ylabel('Nilai')
plt.show()
```

### **Scatter Plot**
ChatGPT Prompt: buatkan kode untuk menampilkan scatter plot untuk semua fitur numerik



```python
# Memilih kolom-kolom numerik
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Menggambar scatter plot untuk setiap pasangan fitur numerik
sns.pairplot(df, kind='scatter', diag_kind='kde')
plt.suptitle('Scatter Plot untuk Fitur Numerik', y=1.02)
plt.show()
```

### **Distribution Plot**
ChatGPT Prompt: buatkan kode untuk distribution dari setiap fitur numerik

```python
# Menggambar histogram atau KDE plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for column in numeric_columns.columns:
    sns.histplot(data=df[column], kde=True, bins=30, alpha=0.5, label=column)

plt.title('Distribusi dari Setiap Fitur Numerik')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

### **Violin Plot**
ChatGPT Prompt: buatkan kode untuk violin plot dari setiap fitur numerik

```python
# Menggambar violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns.columns):
    plt.subplot(3, 4, i + 1)
    sns.violinplot(data=df[column], color='skyblue')
    plt.title(f'Violin Plot {column}')
    plt.xlabel('')

plt.suptitle('Violin Plot untuk Setiap Fitur Numerik', y=1.02)
plt.tight_layout()
plt.show()
```

### **Heat Map**
ChatGPT Prompt: buatkan kode untuk menampilkan heatmap dari fitur - fitur numerik

```python
# Menghitung matriks korelasi antara fitur-fitur numerik
correlation_matrix = df.corr()

# Menggambar heatmap dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap dari Korelasi antara Fitur-Fitur Numerik')
plt.show()
```
