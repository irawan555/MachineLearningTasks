Tentu, berikut adalah langkah-langkah umum untuk melakukan Exploratory Data Analysis (EDA) pada sebuah dataset:

   - Mulailah dengan memahami struktur data, termasuk jumlah kolom dan baris.
   - Periksa header dan beberapa baris pertama dataset untuk mendapatkan gambaran umum tentang data.

   - Analisis statistik deskriptif untuk atribut numerik (seperti mean, median, deviasi standar, dll.).

   - Tangani nilai yang hilang atau null (imputasi nilai atau penghapusan baris/kolom).
   - Tangani outlier jika diperlukan.
   - Lakukan encoding kategori jika ada atribut kategorikal.

   - Analisis univariat untuk setiap atribut, seperti:
     - Untuk atribut kategorikal: Hitung frekuensi dan proporsi setiap kategori.
     - Untuk atribut numerik: Analisis statistik deskriptif, visualisasi distribusi, dan identifikasi outlier.

5. **Analisis Bivariat:**
   - Analisis bivariat untuk memahami hubungan antara dua variabel, seperti:
     - Korelasi antar atribut numerik.
     - Hubungan antara atribut numerik dan kategorikal (box plot, violin plot, dll.).
     - Korelasi antar atribut kategorikal.

6. **Analisis Multivariat:**
   - Analisis multivariat untuk memahami hubungan kompleks antara tiga atau lebih variabel.
   - Gunakan teknik seperti heatmap untuk menunjukkan korelasi antar atribut numerik.

7. **Visualisasi Data:**
   - Gunakan berbagai jenis plot dan grafik untuk memvisualisasikan data.
   - Visualisasikan hubungan antar atribut, distribusi nilai, dan pola-pola yang mungkin tersembunyi dalam data.

8. **Kesimpulan dan Interpretasi:**
   - Tarik kesimpulan dari analisis yang telah dilakukan.
   - Identifikasi pola, tren, anomali, dan insight-insight yang dapat digunakan untuk membuat keputusan lebih lanjut.

9. **Iterasi dan Pembaruan:**
   - Eksplorasi data secara iteratif.
   - Perbarui analisis dan visualisasi berdasarkan wawasan yang diperoleh selama proses.

```python
# Dictionary untuk memetakan nama-nama fitur baru ke nama-nama fitur yang lama
new_column_names = {
    "Total years of experience": "Total experience",
    "Years of experience in Germany": "Experience in Germany",
    "Your main technology / programming language": "Main programming language",
    "Other technologies/programming languages you use often": "Other programming languages",
    "Yearly brutto salary (without bonus and stocks) in EUR": "Yearly brutto salary",
    "Yearly bonus + stocks in EUR": "Yearly bonus and stocks",
    "Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country": "Previous yearly brutto salary",
    "Annual bonus+stocks one year ago. Only answer if staying in same country": "Previous yearly bonus and stocks",
    "Number of vacation days": "Vacation days",
    "Main language at work": "Main work language",
    "Have you lost your job due to the coronavirus outbreak?": "Lost job due to COVID-19",
    "Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week": "Kurzarbeit hours per week",
    "Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR": "Additional monetary support in 2020"
}

# Memperbarui nama-nama fitur
df = df.rename(columns=new_column_names)

# Menampilkan 5 baris teratas setelah memperbarui nama-nama fitur
print(df.head())
```

# **5.Memvisualisasikan Data**

```python
import matplotlib.pyplot as plt

# Memfilter baris berdasarkan rentang usia (20-69)
min_age = 20
max_age = 69

# Menghitung jumlah observasi di setiap kategori usia
bins = [min_age, 30, 40, 50, 60, max_age]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']

df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_counts = df['Age_group'].value_counts()

# Memplot diagram pie
plt.figure(figsize=(8, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Usia (20-69 tahun)')
plt.axis('equal')  # Membuat lingkaran menjadi proporsional
plt.show()
```

GPT Prompt: beri kode untuk menampilkan baris data yang memiliki Age '51-60', '60


```python
# Memfilter baris dengan kelompok usia '51-60' dan '61+'
age_51_to_60 = df.loc[df['Age_group'] == '51-60']
age_60_above = df.loc[df['Age_group'] == '61+']

# Menampilkan data untuk kelompok usia '51-60'
print("Data untuk kelompok usia '51-60':")
print(age_51_to_60)

# Menampilkan data untuk kelompok usia '61+'
print("\nData untuk kelompok usia '61+':")
print(age_60_above)
```
