```python
# Mengimpor fungsi `pipeline` dari pustaka Transformers
from transformers import pipeline

# Membuat pipeline untuk tugas "fill-mask" dengan menggunakan model Camembert
camembert_fill_mask = pipeline("fill-mask", model="camembert-base")

# Mengisi mask di dalam kalimat menggunakan model Camembert
results = camembert_fill_mask("Le camembert est <mask> :)")
```

```python
# Mengimpor kelas-kelas yang diperlukan dari pustaka Transformers untuk pemrosesan bahasa Camembert
from transformers import CamembertTokenizer, CamembertForMaskedLM

# Membuat objek tokenizer untuk bahasa Camembert berdasarkan model yang telah dilatih sebelumnya ("camembert-base")
tokenizer = CamembertTokenizer.from_pretrained("camembert-base")

# Membuat objek model bahasa Camembert untuk tugas pemodelan bahasa yang dilatih sebelumnya ("camembert-base")
model = CamembertForMaskedLM.from_pretrained("camembert-base")
```
